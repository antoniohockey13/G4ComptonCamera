 ### Incident photons hits dispersion in Y-Z

````python
c1 = ROOT.TCanvas()
c1.SetLogy()
file.Hits.Draw('Y>>hY', "DetectorID == 1 && TrackID == 1 && PreKineticEnergy == 70","")
hY = ROOT.gDirectory.Get('hY')
hY.SetTitle('Y/mm')
c1.Draw()

c2 = ROOT.TCanvas()
c2.SetLogy()
file.Hits.Draw('Z>>hZ', "DetectorID == 1 && TrackID == 1 && PreKineticEnergy == 70","")
hZ = ROOT.gDirectory.Get('hZ')
hZ.SetTitle('Z/mm')
c2.Draw()
````

Getting dispersion too, different dispersion spectrum than in the air simulations
